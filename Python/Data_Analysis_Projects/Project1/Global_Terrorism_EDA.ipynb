# Data Cleaning and Preprocessing:

```python
#first we want to clean the data by removing columns with Excessive Null values, removing outliers ....etc
print(df.isna().sum().sort_values(ascending=False))
```

```python
#we want to calculate the percentage of null values in each column and remove any column with percentage above 30
for col in df.columns:
    null_percentage=round((df[col].isna().sum()/df.shape[0])*100)
    if null_percentage>=30:
        print(f"Dropping columns: \"{col}\" as it has Null percentage of ({null_percentage}%)")
        df.drop(col,axis=1,inplace=True)
df.shape
```

```python
#now we return the percentage of missing values in each column
for col in df.columns:
    null_percentage=round((df[col].isna().mean())*100)
    print(f"\"{col}\": ({null_percentage}%)")
    
```

```python
# now we should fill Null values for each column, if the column is categorical we fill it by the mode
#if the column is numerical we fill it by the mean or median depending on the status of the outliers
for col in df.columns:
    if df[col].dtype=="object":
        df[col].fillna(df[col].mode()[0],inplace=True)
    elif df[col].dtype=="int64" or "float64":
        df[col].fillna(df[col].mean(),inplace=True)
```

```python
#now we return the percentage of missing values in each column
for col in df.columns:
    null_percentage=round((df[col].isna().mean())*100)
    print(f"\"{col}\": ({null_percentage}%)")
    
```

**The sns.countplot() method**:
The sns.countplot() function in the Seaborn library is a specialized version of a bar plot that automatically calculates the number of observations (counts) in each categorical bin. It's often the quickest way to visualize the distribution of a single categorical variable.

Here's a brief breakdown of its key parameters and how to use them:

Key sns.countplot() Parameters
Parameter	Purpose	Example Usage
data	The DataFrame or array containing the data.	data=df
x or y	The column name for the categorical variable you want to count. You only need to specify one: x for vertical bars, y for horizontal bars.	x='JobTitle'
hue	The optional categorical variable to group and stack/dodge the counts by. This splits each bar into segments based on a secondary category.	hue='Success Label'
order	A list of strings specifying the exact order in which the bars should appear on the axis.	order=['A', 'B', 'C']
palette	Defines the colors to use for the different levels of the hue variable.	palette='viridis'

```python
#using the count plot 
top_10_countries=df["country_txt"].value_counts().head(10).index #these are the top 10 countries to make the order
plt.figure(figsize=(18,10))
sns.countplot(data=df,x="country_txt",hue="Success Label",order=top_10_countries,palette="viridis")
plt.title("Top 10 countries vs Success Rate")
plt.xlabel("Country")
plt.ylabel("Number of Attacks")
plt.savefig(r'F:\studiess\Machine Learning and Artificial Intellignece\ML&AI_Coding_Academy\ML&AI_Repository\Python\Data_Analysis_Projects\Graphs\top_10_countries_success.png')

plt.show()
```
