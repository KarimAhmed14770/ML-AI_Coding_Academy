## File Handling

*Definition*:File handling is reading from or writing to files using python's built-in functions
*Characterstics*:
1. Files can e opened in different mode:(r) read only, (w) writing,(a) append,b(binary),..etc
2. files should always be closed after use to free system resource
3. with statement is used to automatically handle file closure
4. Exceptions should be handled for cases like file not found
5. supports reading file line by line , reading all the content at once, and writing multiple lines
6. 

```python
#Reading from a file:
#to open a file we use open() function
#its documentation https://docs.python.org/3/library/functions.html#open\
#(with) statement if put before open function, closes the file upon ending the work on it
#it returns a file object that we handle with using alias 
#if it does not find the dile it will give an error (FileNotFoundError)
#syntax: with open(filepath,filemode) as alias_name:
#                    body
#to read from a file we use file_alias_name.read()
#read documentation https://docs.python.org/3/library/io.html#io.TextIOBase.write
#if read takes read(size=-1) if it takes -1 or none it gets all that is in the file
#and stores it in a single string till it finds EOF
#if it takes size=4 it only fetches the first 4 characters in the file
#it gets all the data in the file
#example 
with open('toread.txt','r') as file1:
    file1_content=file1.read()
    print(file1_content)
print()
with open(r"F:\studiess\trials\pythonread.txt",'r') as file2:
    file2_content=file2.read()
    print(file2_content)
```

```python
#to read a line we use readline(size=-1), if size is -1 or none it will return the size of character
#in the line 
with open('toread.txt','r') as file1:
    file1_first_line=file1.readline()
    print(file1_first_line)
    print(type(file1_first_line))
#to read multilple line we use readlines(size=-1), it returns a list of strings with each element
#corresponding to a certain line , if size =-1 or none it gets till EOF 
#if size =4 it gets the line that contains first 4 bytes
#if 30 bytes are in 2 line readlines(30) gets the first two lines
with open('toread.txt','r') as file1:
    file1_lines=file1.readlines()
    print(file1_lines)
    print(type(file1_lines))
```

*File Modes*:
1. Read Mode('r'):opens the file for reading , the file must exist otherwise an error occurs
2. Write Mode('w'):opens the file for writing if the file exist it is emptied, if it doesn't exist it is created
3. Append Mode('a'):opens the file for appending, data is written to the end of the file
4. Binary Mode('b'): can be added to any of the above modes to read or write binary files(images,audio files) ex 'rb' ,'wb','ab'
5. Exclusive creation Mode('x'): opens a file for writing but fails if the file already exists
6. Read and write mode('r+'): opens a file for both read and write

```python
#binary mode for non text files
with open('images.jpeg','rb') as file:
    content=file.read()
    print(content)
```

```python
with open('output_image.jpeg','wb') as im:
    im.write(content)
```

```python
#to check if a file exists using os.path
import os
if os.path.exists('images.jpeg'):
    with open('images.jpeg','rb') as file:
        data=file.read()
        print(data)
else:
    print('file not found')
```

```python
#handling exceptions when using files
try:
    with open('imagesss.jpeg','rb') as file:
        data=file.read()
        print(data)
except FileNotFoundError:
    print("File not found.")
```

## Error and Exception Handling

**Definition**:Exception handling in python allows you to manage and handle runtime errors, preventing the progra from crashing unexpectedly

**Characterstics**:
1. Uses try-except blocs to catch and handle errors.
2. can catch multiple exceptions in a single block.
3. else bloc:executes if no exceptions are raised.
4. finally block:always executes.
5. you can raise custom exceptions using raise.
6. custom exceptions can be created by inheiting from the Exception class.

```python
#basic try-except block:
try:
    num=int(input("enter a number: "))
    print(10/num)
except ZeroDivisionError:
    print("You cannot divide by zero!")
except ValueError:
    print("Invalid input, please enter a number.")
#ZeroDivisionError
#ValueError                                
```

```python
#catching multiple exceptions
try:
    num=int(input("enter a number: "))
    print(10/num)
except (ZeroDivisionError,ValueError) as e:
    print(f"an error occured :{e}")
```

```python
#else block: executed if no exceptions occur
try:
    num=int(input("enter a number: "))
    result=(10/num)
except (ZeroDivisionError,ValueError) as e:
    print(f"an error occured :{e}")
else:
    print(f"result is: {result}")
```

```python
#finally block (mmay be used to close a file after ending)
try:
    file=open('toread.txt','r')
    content=file.read()
except FileNotFoundError:
    print("File not found.")
finally:
     print("This block is always executed.")
     if 'file' in locals():#locals() return all the local variables 
         file.close()
         print("file closed")
    
    
```

```python
#raising exceptions manually
def divide(a,b):
    if(b==0):
        raise ZeroDivisionError("b can't be zero!")
    return a/b

try:
    result=divide(10,0)
except ZeroDivisionError as e:
    print(e)
```

```python
# creating custom exceptions
class NegativeNumberError(Exception):# this makes NegativeNumberError inherits fro class Exception
    pass
def check_positive(number):
    if number<0:
        raise NegativeNumberError("Number cannot be negative!")
    return number

try:
    check_positive(-1)
except NegativeNumberError as e:
    print(e)
```
